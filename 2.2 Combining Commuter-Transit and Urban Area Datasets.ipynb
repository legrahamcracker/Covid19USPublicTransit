#### 3. Merge commuter-transit and urban area data
#### 4. Export combined commuter-transit-urban-area master dataframe

## 1. Importing libraries and data

```python
# import libraries

import pandas as pd
import numpy as np
import os
```

```python
# set up path variable for easy import and export of data

path = '/Users/matthewmacbook/Documents/CareerFoundry/Data Immersion/Achievement 6 - Advanced Analytics and Dashboard Design/COVID-19 Public Transit Project'
```

```python
pd.set_option("display.max_rows", 100)
pd.set_option("display.max_columns", 100)
```

```python
low_memory = False
```

```python
# Import commuter-transit dataset
df_commuter_transit = pd.read_csv(os.path.join(path, 'Datasets', 'Transformed Data', 'commuter_transit_2018_to_2022.csv'))

# Import urban area dataset
df_urban_area = pd.read_csv(os.path.join(path, 'Datasets', 'Clean Data', 'census_urban_area_2020.csv'))

```

## 2. Data Cleaning

```python
df_commuter_transit
```

```python
df_urban_area
```

```python
df_urban_area = df_urban_area.drop(columns = ['Name'])
```

```python
df_urban_area
```

## 3. Merge commuter-transit and urban area data

```python
df_commuter_transit_master = df_commuter_transit.merge(df_urban_area, on='UACE Code', how='inner')
```

```python
df_commuter_transit_master = df_commuter_transit_master.rename(columns = {'Trips': 'Transit Trips', 'Vehicle Miles': 'Transit Vehicle Miles'})
```

```python
df_commuter_transit_master
```

```python
df_commuter_transit_master['UACE Code'] = df_commuter_transit_master['UACE Code'].astype(str).str.zfill(5)
```

```python
df_commuter_transit_master
```

## 4. Export combined commuter-transit-urban-area master dataframe

```python
# Export combined dataframe into csv file in transformed data folder as 'commuter_transit_master_2018_to_2022.csv'
df_commuter_transit_master.to_csv(os.path.join(path, 'Datasets', 'Transformed Data', 'commuter_transit_master_2018_to_2022.csv'), index=False)
```
